```python
#using scikit-learn(sklearn) module to convert labels into numbers
from sklearn.preprocessing import LabelEncoder
```

```python
#encoding columns
le_Brand = LabelEncoder()
le_Year = LabelEncoder()
le_Model = LabelEncoder()
le_UsedOrNew = LabelEncoder()

inputs = df.copy()
```

```python
#learn unique categories and assign number to them
df['Brand_n'] = le_Brand.fit_transform(inputs['Brand'])
df['Year_n'] = le_Brand.fit_transform(inputs['Year'])
df['Model_n'] = le_Brand.fit_transform(inputs['Model'])
df['UsedOrNew_n'] = le_Brand.fit_transform(inputs['UsedOrNew'])
```

```python
# drop all except required encoded inputs_n
inputs_n = inputs.drop(['Brand' , 'Year' , 'Model' , 'UsedOrNew' , 'Transmission' , 'Engine' , 'DriveType' , 'FuelType' , 'Car/Suv' , 'FuelConsumption' , 'Kilometres' , 'ColourExtInt' , 'Price' , 'Location' , 'CylindersinEngine' , 'Doors' , 'Seats' , 'BodyType' , 'Title'], axis = 'columns')
inputs_n
```

```python
#set target as price
target = df['Price']
```

```python
#drop nans from encoded inputs
input_n = inputs_n.dropna()

#drop nans from target
target = target.dropna()
    
#align both(this is the missing part)
common_index = inputs_n.index.intersection(target.index)

inputs_n = inputs_n.loc[common_index]
target = target.loc[common_index]
```

```python
#preparing code to use decision trees from scikit-learn
from sklearn import tree
```

```python
#create the model
model = tree.DecisionTreeClassifier()
```

```python
#train function
model.fit(inputs_n, target)
```

```python
#how good the trained model at predicting target for the giving inputs.
model.score(inputs_n, target)
```
